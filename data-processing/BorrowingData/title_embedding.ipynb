```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

maxlen = 100                               # Cuts off title after 100 words
training_samples = 10000                     # Trains on 200 samples
validation_samples = df_title.shape[0] - training_samples                 # Validates on 10,000 samples
max_words = 10000   
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(df_title.Title)
sequences = tokenizer.texts_to_sequences(df_title.Title)
word_index = tokenizer.word_index

print('Found %s unique tokens.' % len(word_index))
```

```python
import os

glove_dir = '/Users/huangz55/Downloads/'

embeddings_index = {}
f = open(os.path.join(glove_dir, 'glove.6B.100d.txt'), encoding="utf8")
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:])
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

```python
embedding_dim = 100

embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in word_index.items():
    if i < max_words:
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
            
```
